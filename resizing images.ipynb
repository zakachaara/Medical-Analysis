```python
import os
import cv2
```

## adding a border to the initial images to make them square

```python
path = 'B:/archive_tr/training/Images/'
mask = 'B:/archive_tr/training/Masks/'
paths = []
result = []
image_files = sorted(os.listdir(path))
for image_file in image_files :
    paths.append(os.path.join(path, image_file))
mask_files = sorted(os.listdir(mask))
for mask_file in mask_files:
    result.append(os.path.join(mask, mask_file))
print(len(result),len(paths))
```

```python
def read(path):
    if path.endswith(".gif") :
        
    else :
        valid_image = cv2.imread(path)
    
    return valid_image
```

```python
%%time
path = "B:/archive_tr/Train/masks/"
for i in result:
    d = i.split('/')
    r = d[-1].split('.')
    print(r[0])
#     break
    img = read(i)
    shape =img.shape
    if shape == (2336, 3504, 3):
        copy = cv2.copyMakeBorder(img,584,584,0,0,cv2.BORDER_CONSTANT, value = 0)
    elif shape == (584, 565, 3):
        copy = cv2.copyMakeBorder(img,0,0,10,9,cv2.BORDER_CONSTANT, value = 0)
    else :
        copy = cv2.copyMakeBorder(img,1,0,20,20,cv2.BORDER_CONSTANT, value = 0)
    cv2.imwrite(path + r[0]+'.jpg' ,cv2.cvtColor(copy , cv2.COLOR_BGR2RGB) )
#     break
    
```
